```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_ratios = [2, 4, 6, 8, 10]

paths_sup = [
    
    f'checkpoints/sup_{ratio}' for ratio in data_ratios
]

paths_sup_nf = [
    f'checkpoints/sup_{ratio}_nf' for ratio in data_ratios
]

paths_unsup = [
    f'checkpoints/unsup_{ratio}' for ratio in data_ratios
]


paths_unsup_nf = [
    f'checkpoints/unsup_{ratio}_nf' for ratio in data_ratios
]
```

```python
paths_unsup_nf
```

```python
sup_values = []
unsup_values = []

sup_values_nf = []
unsup_values_nf = []

for path in paths_sup:
    sup_values.append(np.load(f"{path}/test_auroc_mean.npy")[0])

for path in paths_unsup:
    unsup_values.append(np.load(f"{path}/test_auroc_mean.npy")[0])
    
for path in paths_sup_nf:
    sup_values_nf.append(np.load(f"{path}/test_auroc_mean.npy")[0])

for path in paths_unsup_nf:
    unsup_values_nf.append(np.load(f"{path}/test_auroc_mean.npy")[0])

    
```

```python
colors = ['#8a2244', '#da8c22', '#c687d5', '#80d6f8', '#440f06', '#000075', '#000000']
fig = plt.figure(figsize=(8, 7))

plt.title('Without No Finding')


plt.plot(sup_values, 'b', marker='x', label = 'Sup')
plt.plot(unsup_values, 'g',marker='D', label = 'Semi Sup')

plt.legend(loc = 'lower right', )
plt.ylabel('AUROC Mean')
plt.xlabel('% Labelled Examples')
plt.xticks(range(len(data_ratios)), data_ratios)
fig.savefig(f'plots/Comparison.pdf')
```

```python
fig = plt.figure(figsize=(8, 7))

# plt.title('Receiver Operating Characteristic')
plt.plot(sup_values_nf, 'b', marker='x', label = 'Sup')
plt.plot(unsup_values_nf, 'g',marker='D', label = 'Semi Sup')
plt.title('With No Finding')

plt.legend(loc = 'lower right', )
plt.ylabel('AUROC Mean')
plt.xlabel('% Labelled Examples')
plt.xticks(range(len(data_ratios)), data_ratios)

fig.savefig(f'plots/Comparison_nf.pdf')
```
